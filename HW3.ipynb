```python
#使用函數的方式 抓取DATA放進DICTIONARY之中
def get_stu_data(data):
    result = []
    
    for ele in data:
        title = ele.select("div.title")[0].text.strip()
        if "刪除" in title :#過濾不適合的標題
            continue
        #print("標題:",title)
        time=ele.select("div.date")[0].text
        #print("時間:",time)
        writter=ele.select("div.author")[0].text
        #print("作者:",writter)
        raw_link=ele.select("div.title a")[0]["href"]
        doma_name="https://ptt.cc"
        link=doma_name + raw_link
        #print("連結:",link)
        #print("*"*100)
        
        result.append({
            "title": title,
            "time": time,
            "Author": writter,
            "link" : link     
        })
    return result
    
get_stu_data(data)
```

```python
#使用函數的方式 抓取DATA放進DICTIONARY之中
def get_stu_data(soup):
    
    data=soup.select("div.r-ent")
    result = []
    
    for ele in data:
        title = ele.select("div.title")[0].text.strip()
        if "刪除" in title :#過濾不適合的標題
            continue
        #print("標題:",title)
        time=ele.select("div.date")[0].text
        #print("時間:",time)
        writter=ele.select("div.author")[0].text
        #print("作者:",writter)
        raw_link=ele.select("div.title a")[0]["href"]
        doma_name="https://ptt.cc"
        link=doma_name + raw_link
        #print("連結:",link)
        #print("*"*100)
        
        result.append({
            "title": title,
            "time": time,
            "Author": writter,
            "link" : link     
        })
    return result
    
get_stu_data(soup)

###發request取得原始碼
from bs4 import BeautifulSoup as bs
import requests

url="https://www.ptt.cc/bbs/studyteacher/index.html"
headers={
    
    "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
#模擬瀏覽器如程式一般
}

res = requests.get(url,headers = headers)
soup=bs(res.text,"lxml")

##解析首頁資料
output=[]
result=get_stu_data(soup)
output+=result
```

```python
url="https://www.ptt.cc/bbs/studyteacher/index.html"
headers={
    
    "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
#模擬瀏覽器如程式一般
}

res = requests.get(url,headers = headers)

soup=bs(res.text,"lxml")
##解析首頁資料
output=[]
result=get_stu_data(soup)
output+=result

#抓取分頁資料
up_page=soup.select("div#action-bar-container div.btn-group-paging a")[1]["href"]#打開標籤每層的打開 可以看原始碼
page_num=(up_page.replace("/bbs/studyteacher/index","").replace(".html",""))#使用取代的方式只抓取中間的數字
print(page_num)
for i in range(5):#從首頁開始遞減到5去抓取5個頁碼以下的資料
    url="https://www.ptt.cc/bbs/studyteacher/index{}.html".format(int(page_num)-i)
    res = requests.get(url,headers = headers)
    soup=bs(res.text,"lxml")
    
    result=get_stu_data(soup)
    output+=result
    
    print("{} is ok.".format(url))
print("Done.")
```
